```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import skimage.io as io
import skimage.transform as trs
import skimage.color as sColor
from bokeh.plotting import figure, show, output_file
from bokeh.models import HoverTool
from PIL import Image
```

```python
images = io.imread_collection("../annunc2/*.jpg")
plt.figure(figsize=(10,10))

plt.imshow(images[0])
```

```python


sh = (images[0].shape[0], images[0].shape[1],1)
onesImg = np.ones(sh)*255.0

newImg = np.concatenate((images[0],onesImg),axis =2)

#newImg = newImg[:,:,[2,1,0,3]]
```

```python
plt.figure(figsize=(10,10))
plt.imshow(newImg.astype(np.uint8))
```

```python
plt.hist(images[0].ravel())
```

```python
image= (newImg).astype(np.uint8)
p = figure(x_range=(0,2000), y_range=(0, 2000),sizing_mode = 'stretch_both')
          # tools=[HoverTool(tooltips=[("x", "$x"), ("y", "$y"), ("value", "@image")])])
p.image_rgba(image=[image], x=0, y=0, dw=sh[0]/2, dh=sh[1]/2)
show(p,notebook_handle=True)
```

```python
N = 20
img = np.empty((N,N), dtype=np.uint32)
view = img.view(dtype=np.uint8).reshape((N, N, 4))
for i in range(N):
    for j in range(N):
        view[i, j, 0] = int(i/N*255)
        view[i, j, 1] = 158
        view[i, j, 2] = int(j/N*255)
        view[i, j, 3] = 255

p = figure(x_range=(0,10), y_range=(0,10))

# must give a vector of images
p.image_rgba(image=[img], x=0, y=0, dw=10, dh=10)

show(p,notebook_handle=True)  # open a browser
```
