```python
data= {"x_tsne":[1,2,3,4,5],"y_tsne":[2,3,5,4,8],"hue":[2,2,3,8,7],"sat":[5,8,1,2,3]}
data=pd.DataFrame(data=data)
columns = sorted(data.columns)

def create_figure(x,y):
    xs = data[x].values
    ys = data[y].values
    x_title = x
    y_title = y


    p = figure(plot_height=600, plot_width=800, tools='pan,box_zoom,hover,reset', **kw)
    p.xaxis.axis_label = x_title
    p.yaxis.axis_label = y_title

    #p.image(image = images.extract_lowRes()[:5],x=xs, y=ys, dw=32,dh=32)
    p.circle(x=xs, y=ys, size=35)
    push_notebook()

    
output_notebook()
xs = data["hue"].values
ys = data["sat"].values
x_title = 'hue'
y_title = 'sat'
kw = {}
kw['title'] = "%s vs %s" % (x_title, y_title)
p = figure(plot_height=600, plot_width=800, tools='pan,box_zoom,hover,reset', **kw)
p.xaxis.axis_label = x_title
p.yaxis.axis_label = y_title
p.circle(x=xs, y=ys, size=35)

show(p,notebook_handle = True)

```

```python
from ipywidgets import interact
import numpy as np

from bokeh.io import push_notebook, show, output_notebook
from bokeh.plotting import figure
output_notebook()
```

```python
x = np.linspace(0, 2*np.pi, 2000)
y = np.sin(x)
```

```python
p = figure(title="simple line example", plot_height=300, plot_width=600, y_range=(-5,5))
r = p.line(x, y, color="#2222aa", line_width=3)
```

```python
def update(f, w=1, A=1, phi=0):
    if   f == "sin": func = np.sin
    elif f == "cos": func = np.cos
    elif f == "tan": func = np.tan
    r.data_source.data['y'] = A * func(w * x + phi)
    push_notebook()
```

```python
show(p, notebook_handle=True)
```

```python
interact(update, f=["sin", "cos", "tan"], w=(0,100), A=(1,5), phi=(0, 20, 0.1))

```
